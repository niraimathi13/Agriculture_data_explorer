```python
#checking duplicate rows
duplicate_rows=df[df.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")
```

```python
#checking duplicate columns
duplicate_columns=df.columns[df.columns.duplicated()]
print("Duplicate column names:",duplicate_columns.tolist())
```

```python
# Check if there are any negative values
has_negative_values=(df.select_dtypes(include='number')<0).any().any()
if has_negative_values:
    print("There are negative values in the dataset")
else:
    print("No negative values found in the dataset")
```

```python
#columns containing negative values
neg_columns=(df.select_dtypes(include='number')<0).any()
columns_with_negatives=neg_columns[neg_columns].index.tolist()
print("Columns containing negative values:")
for col in columns_with_negatives:
    print(f"- {col}")
```

```python
#negative values replaced with zero
df[df.select_dtypes(include='number').columns]=df.select_dtypes(include='number').map(lambda x: 0 if x<0 else x)
print("All negative values have been replaced with 0")
```

```python
converted_columns = {}
for col in df.columns:
    if "1000 ha" in col:
        new_column=col.replace(" (1000 ha)", " (ha)")
        converted_columns[new_column]=df[col]*1000
    elif "1000 tons" in col:
        new_column=col.replace(" (1000 tons)", " (kg)")
        converted_columns[new_column]=df[col]*1000000
# Adding all converted columns at once
df = pd.concat([df, pd.DataFrame(converted_columns)],axis=1)
# Droping the original unit columns
cols_to_drop = [col for col in df.columns if "1000 ha" in col or "1000 tons" in col]
df.drop(columns=cols_to_drop, inplace=True)
```

```python
# SQL QUERY
#(1) Year-wise trend of rice production across states(top 3)
mycursor.execute("""with top_states as (select `State Name` from agriculture.data group by `State Name` order by sum(`RICE PRODUCTION (kg)`)desc limit 3)
select `Year`,`State Name`,sum(`RICE PRODUCTION (kg)`) as total_rice_production from agriculture.data where `State Name` in (select `State Name` from top_states) group by `Year`,`State Name` order by `Year`,`State Name`""")
a=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(a, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
#(2) Top 5 districts by wheat yield increase over the last 5 years
mycursor.execute("""with wheat_yield as(select `Dist Name` as district,Year,sum(`WHEAT PRODUCTION (kg)`)/nullif(sum(`WHEAT AREA (ha)`),0) as yield from agriculture.data where Year in((select max(Year) from agriculture.data),(select max(Year) from agriculture.data)-5)group by `Dist Name`,Year),
pivoted as (select district,max(case when Year=(select max(Year) from agriculture.data) then yield end) as recent_yield,max(case when Year=(select max(Year) from agriculture.data)-5 then yield end) as past_yield from wheat_yield group by district),
yield_diff as (select district,recent_yield,past_yield,(recent_yield - past_yield) as yield_increase from pivoted where recent_yield is not null and past_yield is not null)
select * from yield_diff order by yield_increase desc limit 5""")
a=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(a, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
#(3) States with the highest growth in oilseed production (5-year growth rate)
mycursor.execute("""with oilseed_totals as(select `State Name` as state,Year,sum(`OILSEEDS PRODUCTION (kg)`) as total_production from agriculture.data where Year in((select max(Year) from agriculture.data),(select max(Year) from agriculture.data)-5)group by `State Name`,Year),
pivoted as (select state,max(case when Year=(select max(Year) from agriculture.data) then total_production end) as recent_prod,max(case when Year=(select max(Year) from agriculture.data)-5 then total_production end) as past_prod from oilseed_totals group by state),
growth_calc as (select state,recent_prod,past_prod,((recent_prod-past_prod)/past_prod)*100 as growth_rate from pivoted where recent_prod is not null and past_prod is not null and past_prod!=0)
select * from growth_calc order by growth_rate desc limit 5""")
a=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(a, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
#(4) District-wise correlation between area and production for major crops(rice,wheat,maize)
mycursor.execute("""with rice_stats as(select `Dist Name` as district,count(*) as n,sum(`RICE AREA (ha)`) as sum_x,sum(`RICE PRODUCTION (kg)`) as sum_y,sum(`RICE AREA (ha)`*`RICE PRODUCTION (kg)`) as sum_xy,sum(power(`RICE AREA (ha)`,2)) as sum_x2,sum(power(`RICE PRODUCTION (kg)`,2)) as sum_y2 from agriculture.data where `RICE AREA (ha)` is not null and `RICE PRODUCTION (kg)` is not null group by `Dist Name`),
rice_corr as (select district,'Rice' as crop,(n*sum_xy-sum_x*sum_y)/(sqrt(n*sum_x2-power(sum_x,2))*sqrt(n*sum_y2-power(sum_y,2))) as correlation from rice_stats where (n*sum_x2-power(sum_x,2))>0 and (n*sum_y2-power(sum_y,2))>0),
wheat_stats as (select `Dist Name` as district,count(*) as n,sum(`WHEAT AREA (ha)`) as sum_x,sum(`WHEAT PRODUCTION (kg)`) as sum_y,sum(`WHEAT AREA (ha)`*`WHEAT PRODUCTION (kg)`) as sum_xy,sum(power(`WHEAT AREA (ha)`,2)) as sum_x2,sum(power(`WHEAT PRODUCTION (kg)`,2)) as sum_y2 from agriculture.data where `WHEAT AREA (ha)` is not null and `WHEAT PRODUCTION (kg)` is not null group by `Dist Name`),
wheat_corr as (select district,'Wheat' as crop,(n*sum_xy-sum_x*sum_y)/(sqrt(n*sum_x2-power(sum_x,2))*sqrt(n*sum_y2-power(sum_y,2))) as correlation from wheat_stats where (n*sum_x2-power(sum_x,2))>0 and (n*sum_y2-power(sum_y,2))>0),
maize_stats as (select `Dist Name` as district,count(*) as n,sum(`MAIZE AREA (ha)`) as sum_x,sum(`MAIZE PRODUCTION (kg)`) as sum_y,sum(`MAIZE AREA (ha)`*`MAIZE PRODUCTION (kg)`) as sum_xy,sum(power(`MAIZE AREA (ha)`,2)) as sum_x2,sum(power(`MAIZE PRODUCTION (kg)`,2)) as sum_y2 from agriculture.data where `MAIZE AREA (ha)` is not null and `MAIZE PRODUCTION (kg)` is not null group by `Dist Name`),
maize_corr as (select district,'Maize' as crop,(n*sum_xy-sum_x*sum_y)/(sqrt(n*sum_x2-power(sum_x,2))*sqrt(n*sum_y2-power(sum_y,2))) as correlation from maize_stats where (n*sum_x2-power(sum_x,2))>0 and (n*sum_y2-power(sum_y,2))>0)
select * from rice_corr union all select * from wheat_corr union all select * from maize_corr order by district,crop""")
a=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(a, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
#(5) Yearly production growth of cotton in top 5 cotton producing states
mycursor.execute("""with top_states as (select `State Name` from agriculture.data group by `State Name` order by sum(`COTTON PRODUCTION (kg)`) desc limit 5),
cotton_production as(select `State Name` as state,Year,sum(`COTTON PRODUCTION (kg)`) as total_production from agriculture.data where `State Name` in(select `State Name` from top_states) group by `State Name`,Year),
growth_calc as(select state,Year,total_production,lag(total_production) over(partition by state order by Year)as prev_year_production from cotton_production)
select state,Year,total_production,prev_year_production,round((total_production-prev_year_production)/prev_year_production*100,2)as growth_percent from growth_calc where prev_year_production is not null order by state,Year""")
a=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(a, headers=[i[0] for i in mycursor.description], tablefmt='psql'))

```
